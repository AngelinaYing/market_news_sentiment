```python
import json
import pandas as pd
```

```python
def json_to_df(json_file_path, dataframe):
    with open(json_file_path, 'r') as file:
        content = json.load(file)["response"]["docs"]
    df = pd.json_normalize(content)
    return pd.concat([df, dataframe], ignore_index=True)
```

```python
column_names = ['abstract', 'web_url', 'snippet', 'lead_paragraph', 'source', 'multimedia', 'keywords', 'pub_date', 'document_type', 'news_desk','section_name', 'type_of_material', '_id', 'word_count', 'uri', 'headline.main', 'headline.kicker', 'headline.content_kicker', 'headline.print_headline', 'headline.name', 'headline.seo', 'headline.sub', 'byline.original', 'byline.person', 'byline.organization', 'print_section', 'print_page', 'subsection_name']
df = pd.DataFrame(columns=column_names)
```

```python
df = json_to_df("json/archive_2019_01.json", df)
df = json_to_df("json/archive_2019_02.json", df)
df = json_to_df("json/archive_2019_03.json", df)
df = json_to_df("json/archive_2019_04.json", df)
df = json_to_df("json/archive_2019_05.json", df)
df = json_to_df("json/archive_2019_06.json", df)
df = json_to_df("json/archive_2019_07.json", df)
df = json_to_df("json/archive_2019_08.json", df)
df = json_to_df("json/archive_2019_09.json", df)
df = json_to_df("json/archive_2019_10.json", df)
df = json_to_df("json/archive_2019_11.json", df)
df = json_to_df("json/archive_2019_12.json", df)
```
